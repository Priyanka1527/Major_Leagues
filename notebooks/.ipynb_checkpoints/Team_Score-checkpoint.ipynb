```python
#handling the missing values using fillna()
matches = matches.fillna(matches['score1'].value_counts().index[0])
matches = matches.fillna(matches['score2'].value_counts().index[0])
matches = matches.fillna(matches['xg1'].value_counts().index[0])
matches = matches.fillna(matches['xg2'].value_counts().index[0])
matches = matches.fillna(matches['nsxg1'].value_counts().index[0])
matches = matches.fillna(matches['nsxg2'].value_counts().index[0])
matches = matches.fillna(matches['adj_score1'].value_counts().index[0])
matches = matches.fillna(matches['adj_score2'].value_counts().index[0])
```

```python
#no. of null values after mode imputation
print(matches.isnull().values.sum())
```

```python
#No. of unique leagues = 37
print(matches['league'].value_counts().count())
```

```python
#mathches by league
plt.figure(figsize=(8,8))
ax = sns.countplot(y = matches["league"],
                   order=matches["league"].value_counts().index,
                   linewidth = 1,
                   edgecolor = "k"*matches["league"].nunique()
                 )
for i,j in enumerate(matches["league"].value_counts().values):
    ax.text(.7,i,j,weight = "bold")
plt.title("Matches by league")
plt.show()
```

```python
team_list = matches["team1"].tolist()
teams = set(team_list)
teams = list(teams)
len(teams)
```

```python
#teams with highest scores
h_t = matches.groupby("team1")["score1"].sum().reset_index()
a_t = matches.groupby("team2")["score2"].sum().reset_index()
h_t = h_t.sort_values(by="team1",ascending= False)
a_t = a_t.sort_values(by="team2",ascending= False)

x = h_t
x = x.rename(columns={'team1':"team", 'score1':"goals"})
y = a_t
y = y.rename(columns={'team2':"team", 'score2':"goals"})

goals = pd.concat([x,y])
goals = goals.groupby("team")["goals"].sum().reset_index().sort_values(by = "goals",ascending = False)
plt.figure(figsize=(9,14))
ax = sns.barplot(x="goals",y="team",
                  data=goals[:30],palette="rainbow",
                 linewidth = 1,edgecolor = "k"*30)
for i,j in enumerate(goals["goals"][:30]):
    ax.text(.3,i,j,weight="bold",color = "k",fontsize =12)
plt.title("Teams with highest total goals ")
plt.show()
```

```python
#We don't need date, League ID, league name and team mnames
matches = matches.drop('league_id', axis=1)
```

```python
matches = matches.drop('league', axis=1)
```

```python
matches = matches.drop('team2', axis=1)
```

```python
matches = matches.drop('spi2', axis=1)
matches = matches.drop('prob2', axis=1)
matches = matches.drop('proj_score2', axis=1)
matches = matches.drop('importance2', axis=1)
matches = matches.drop('score2', axis=1)
matches = matches.drop('xg2', axis=1)
matches = matches.drop('nsxg2', axis=1)
matches = matches.drop('adj_score2', axis=1)
```

```python
matches.head()
```

```python
matches = matches.drop('team1', axis=1)
```

```python
#Split dataset into training and test
X = matches.drop('score1', axis=1)
y = matches[['score1']]

from sklearn.model_selection import train_test_split

# Split X and y into X_
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
#Train the model
from sklearn.linear_model import LinearRegression

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
#Now that our model is trained, we can view the coefficients of the model using regression_model.coef_, 
#which is an array of tuples of coefficients.
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
#regression_model.intercept_ contains an array of intercepts (Î²0 values)
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
#Note that, because weâ€™ve not done any feature scaling or dimensionality reduction, 
#we canâ€™t say anything about the relative importance of each of our features given these coefficients 
#because the features are not of the same scale.
```

```python
#Scoring model
regression_model.score(X_test, y_test)
```

```python
#So in our model, 60.37% of the variability in Y can be explained using X
```

```python
#We can also get the mean squared error using scikit-learnâ€™s mean_squared_error method 
#and comparing the prediction for the test data set with the ground truth for the data test set:
```

```python
from sklearn.metrics import mean_squared_error

y_predict = regression_model.predict(X_test)

regression_model_mse = mean_squared_error(y_predict, y_test)

regression_model_mse
```

```python
import math

math.sqrt(regression_model_mse)
```

```python
#So we are an average of 0.68 score away from the ground truth score when making predictions on our test set.
```
